**Dataset with missing values**

```python
df2=pd.read_csv('/content/gdrive/MyDrive/brittleness-index.csv')
```

```python
df2.head()
```

```python
df2.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(df2)
dff = imputer.transform(df2)
```

```python
df2.shape
imputer.statistics_
```

```python
dff=pd.DataFrame(dff,columns=df2.columns)
```

```python
dff.isna().sum()
```

```python
dff.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
transpip=Pipeline([('std_scaler',StandardScaler()),])
dfff=transpip.fit_transform(dff)
```

```python
dffff=pd.DataFrame(dfff,columns=dff.columns)
dffff.head()
```
